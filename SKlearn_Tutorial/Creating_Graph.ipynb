# Creating Graphs

## Import matplotlib.pyplot

```python
import matplotlib.pyplot as plt
```

## Creating Datapoints

```python
x = [i for i in range(10)]
y = [2*i for i in range(10)]
print(x)
print(y)
```

## Simple Plotting

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
```

## Scattering

```python
plt.scatter(x,y)
```
